```python
def drop_duplicate_records(dataset_org):
    duplicate_records = dataset_org.duplicated()
    print("Duplicate records count:",sum(duplicate_records))
    hr_dataset_unique = dataset_org.drop_duplicates()
    print("After dropping duplicates:",hr_dataset_unique.shape)
    return hr_dataset_unique
```

```python
hr_dataset = drop_duplicate_records(hr_dataset)
hr_dataset.shape
```

```python
def find_column_duplicates(dataset_org, column):
    return dataset_org.duplicated(column, keep = False)
```

```python
empid_duplicates = find_column_duplicates(hr_dataset, 'EmpID')
empid_duplicates.sum()
```

```python
def drop_columns(data_set, columns):
    return data_set.drop(columns, axis = 1, inplace = True) 
    
```

```python
import datetime
from datetime import date

def calculate_year(column, dateFormat):
    today = date.today()
    difference = today.year - hr_dataset[column].apply(lambda x: datetime.datetime.strptime(x,dateFormat).date().year)
    return difference
```

```python
hr_dataset['Age'] = calculate_year('DOB', "%d-%m-%Y")
```

```python
hr_num_ds = hr_dataset[['Age','YearsOfExperience','PayRate','EngagementSurvey','EmpSatisfaction','SpecialProjectsCount']].values
```

```python
hr_num_ds = pd.DataFrame(hr_num_ds)
```

```python
hr_num_ds.head()
```

```python
from sklearn import preprocessing 
  

""" MIN MAX SCALER """
  
min_max_scaler = preprocessing.MinMaxScaler(feature_range =(0, 1)) 
  
# Scaled feature 
x_after_min_max_scaler = min_max_scaler.fit_transform(hr_num_ds) 
  
print ("\nAfter min max Scaling : \n", x_after_min_max_scaler) 
  
  
""" Standardisation """
  
Standardisation = preprocessing.StandardScaler() 
  
# Scaled feature 
x_after_Standardisation = Standardisation.fit_transform(hr_num_ds) 
  
print ("\nAfter Standardisation : \n", x_after_Standardisation) 
```
